Assignment 1: Temperature Check

```python
temperature = int(input("Enter current temperature: "))

if temperature < 10:
    print("It's too cold!")
elif temperature > 10 and temperature < 25:
    print("It's cool outside!")
else:
    print("It's hot outside!")
```

Assignment 2: ATM Withdrawal System

```python
balance = int(input("Enter your balance: "))
withdraw = int(input("Enter your withdraw: "))

if balance < withdraw:
    print("Insufficient Balance")
elif balance >= withdraw:
    print("Transaction Successful")
else:
    remaining = balance - withdraw
    print(f"Remaining_balance: {remaining}")
```

Assignment 3: Student Grade Calculator

```python
marks = int(input("Enter your marks(0-100): "))

if marks >= 90 and marks <= 100:
    print("You have got A+")
elif marks >= 80 and marks <= 89:
    print("You have got A")
elif marks >= 70 and marks <= 79:
    print("You have got B")
elif marks >= 60 and marks <= 69:
    print("You have got C")
elif marks < 60:
    print("You have Failed!")
```

Assignment 4: Bus Ticket Discount

```python
age = int(input("Enter your age: "))

if age < 5:
    discount = "Free"
elif age > 5 and age < 18:
    discount = age * 50 / 100
elif age > 60:
    discount = age * 30 / 100
else:
    discount = "Full price"
    
print(f"{discount}")
```

Assignment 5: Restaurant Bill with Discount

```python
bill = int(input("Enter your bill: "))

if bill > 1000:
    amount = bill * 10 / 100
else:
    amount = "No discount"

print(f"{amount}")
```

Assignment 7: Traffic Light System

```python
color = input("Enter traffic light(red, yellow or green): ")

if color == "red":
    print("Stop the Car")
elif color == "yellow":
    print("Ready for going")
elif color == "green":
    print("Go from here")
else:
    print("Invalid color")
```

Assignment 8: Leap Year Checker

```python
year = int(input("Enter any year: "))

if year % 400 == 0:
    print("Leap year")
elif year % 4 == 0 and year % 100 != 0:
    print("leap year")
else:
    print("Is not leap year")
```

Assignment 10: Electricity Bill Calculator

```python
units = int(input("Enter your current units: "))

if units < 100:
    bill = units * 5
elif units >= 101 and units <= 200:
    bill = (units - 100) * 7 
else:
    bill = (units - 200) * 10
print(f"{bill}")
```
